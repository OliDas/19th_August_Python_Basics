#### Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class bank :
    
    def __init__(self, accnt_no, accnt_nm, bal):
        self.account_number = accnt_no
        self.account_name = accnt_nm
        self.balance = bal
        
    def withdraw(self, wthdrw_amnt):
        self.balance -= wthdrw_amnt
        print('\nRs.',wthdrw_amnt,'has been withdrawn from you account.')
        
    def deposit(self, depo_amnt):
        self.balance += depo_amnt
        print('\nRs.',depo_amnt,'has been deposited to your account.')
        
    def showDetails(self):
        print("\nAccount Number :",self.account_number)
        print("Account holder Name :",self.account_name)
        print("Bank Balance :",self.balance)
        
        
user = bank(987125,'Nishant',5000)
ch = 0
while ch!=4 :
    ch = int(input("\n1. Show Details \n2. Withdraw \n3. Deposit \n4. Exit\nEnter your choice : "))
    if(ch==1):
        user.showDetails()
    elif(ch==2):
        wthdrw_amnt = int(input("\nEnter amount to be withdrawn : "))
        user.withdraw(wthdrw_amnt)
    elif(ch==3):
        depo_amnt = int(input("\nEnter amount to be deposited : "))
        user.deposit(depo_amnt)
    
```

#### Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
class employee :
    
    def __init__(self, eid, nm, sal):
        self.emp_id = eid
        self.name = nm
        self.salary = sal
        
    def employeeDetails(self):
        print("\nEmployee ID =",self.emp_id)
        print("Employee Name =",self.name)
        print("Salary =",self.salary)
        
    def bonus(self, rating):
        if(rating==5):
            bonus=0.05*self.salary
        elif(rating==4):
            bonus=0.04*self.salary
        elif(rating==3):
            bonus=0.03*self.salary
        elif(rating==2):
            bonus=0.02*self.salary
        else:
            bonus=0.01*self.salary
        print("Your yearly bonus =",bonus)
        self.salary+=bonus
            

emp = employee(713,'Kakashi',1000000)
emp.employeeDetails()
rating = int(input("\nPlease enter your this year's rating : "))
emp.bonus(rating)
emp.employeeDetails()
```

#### Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class VehicleRental :
    
    def __init__(self):
        self.available = {'Bike':4, 'Scooty':16, 'Sedan':13, 'Hatchback':7, 'SUV':5}
        
    def vehicleRent(self, vehicle, qty):
        if(self.available[vehicle]-qty>=0):
            self.available[vehicle]-=qty
            print()
            print(qty,vehicle,'rented.')
            self.vehicleAvaiable()
        else:
            print("\nSorry, enough",vehicle,'not present at the moment.')
    
    def vehicleReturn(self, vehicle, qty):
        self.available[vehicle]+=qty
        print()
        print(qty,vehicle,'returned.')
        self.vehicleAvaiable()
        
    def vehicleAvaiable(self):
        print("\nAvailable vehicles : ")
        for i in self.available.items():
            print(i)
    
        
user = VehicleRental()
user.vehicleAvaiable()
ch=0
while(ch!=3):
    ch=int(input("\nOptions : \n1. Rent \n2. Return \n3. Exit \nChoice : "))
    if(ch==1):
        print("\nOptions :")
        for i in user.available.keys():
            print(i)
        v=input("Enter the type of vehicle you want to rent : ")
        q=int(input("Enter the quantity : "))
        user.vehicleRent(v,q)
    elif(ch==2):
        print("\nOptions :")
        for i in user.available.keys():
            print(i)
        v=input("Enter the type of vehicle you want to return : ")
        q=int(input("Enter the quantity : "))
        user.vehicleReturn(v,q)
```

#### Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class library :
    
    def __init__(self):
        self.all_books = {'DSA':10, 'OS':10, 'Python':15, 'ML':5, 'DL':5}
        
    def displayBooks(self): 
        print("\nAvailable books : ")
        for i in self.all_books.items():
            print(i)
            
    def addNewBook(self, book, count):
        self.all_books[book]=count
        print()
        print(count,book,"books added to the stock.")
        self.displayBooks()
        
    def returnBook(self, book, count):
        self.all_books[book]+=count
        print()
        print(count,book,'books returned.')
        self.displayBooks
              
    def borrowBook(self, book, count):
        if(self.all_books[book]-count>=0):
            self.all_books[book]-=count
            print()
            print(count,book,'borrowed.')
            self.displayBooks()
        else:
            print("\nWe don't have enough",book,'books at the moment.')
            

user = library()
ch=0
while(ch!=5):
    ch=int(input("\nOptions : \n1. View all books \n2. Borrow books \n3. Return Books \n4. Add new books \n5. Exit \nChoice : "))
    if(ch==1):
        user.displayBooks()
    elif(ch==2):
        b=input("Enter the book to be borrowed : ")
        n=int(input("Quantity : "))
        user.borrowBook(b,n)
    elif(ch==3):
        b=input("Enter the book to be returned : ")
        n=int(input("Quantity : "))
        user.returnBook(b,n)
    elif(ch==4):
        b=input("Enter the book to be added to the stock : ")
        n=int(input("Quantity : "))
        user.addNewBook(b,n)
        
```

#### Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class Inventory:
    
    def __init__(self):
        self.products = {'pen':20, 'pencil':15, 'eraser':10, 'sharpener':5}
        
    def displayProducts(self):
        print("\nAvailable Products : ")
        for i in self.products.items():
            print(i)
            
    def addNewProducts(self, product, qty):
        self.products[product]=qty
        print()
        print(qty,product,'added to the stock.')
        self.displayProducts()
        
    def updateProductQty(self, product, qty, option):
        if(option==1):
            self.products[product]+=qty
            print()
            print(qty,'more',product,'added to the current stock.')
            self.displayProducts()
        else:
            if(self.products[product]-qty>=0):
                self.products[product]-=qty
                print()
                print(qty,product,'sold.')
                self.displayProducts()
            else:
                print("Not enough",product,'is there in stock at the moment.')
            
            
user = Inventory()
ch=0
while(ch!=4):
    ch=int(input("\nOptions : \n1. Add new products \n2. Update old products quantity \n3. Display all products \n4. Exit \nChoice : "))
    if(ch==1):
        prd = input("Enter the new product to be added : ")
        qty = int(input("Enter quantity : "))
        user.addNewProducts(prd,qty)
    elif(ch==2):
        option = int(input("\n1. Re-stock products \n2. Sell products \nChoice :"))
        prd = input("Enter the product to update it's quantity : ")
        qty = int(input("Enter quantity : "))
        user.updateProductQty(prd,qty,option)
    elif(ch==3):
        user.displayProducts()
```

#### Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class shape:
    
    def __init__(self, l=0, w=0, h=0):
        self.length=l
        self.width=w
        self.height=h
        
    def square(self):
        peri = 4*self.length
        area = self.length*self.length
        print("Perimeter of the square =",peri)
        print("Area of the square =",area)
        
    def rectangle(self):
        peri = 2 * (self.length+self.width)
        area = self.length * self.width
        print("Perimeter of the rectangle =",peri)
        print("Area of the rectangle =",area)
    
    def cube(self):
        peri = 12*self.length
        area = 6 * self.length * self.length
        print("Perimeter of the cube =",peri)
        print("Area of the cube =",area)
        
    def cuboid(self):
        peri = 4 * (self.length + self.width + self.height)
        area = 2 * (self.length*self.width + self.width*self.height + self.height*self.length)
        print("Perimeter of the cuboid =",peri)
        print("Area of the cuboid =",area)
        



print("\nEnter the side of the square : ")
x=int(input())
s1=shape(x)
s1.square()

print("\nEnter the sides of the rectangle : ")
l=[]
for i in range(2):
    x=int(input())
    l.append(x)
s2=shape(l[0],l[1])
s2.rectangle()

print("\nEnter the side of the cube : ")
x=int(input())
s3=shape(x)
s3.cube()    
    
print("\nEnter the sides of the cuboid : ")
l=[]
for i in range(3):
    x=int(input())
    l.append(x)
s4=shape(l[0],l[1],l[2])
s4.cuboid()
```

#### Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class Student:
    
    def __init__(self, ID, name, eng, mat, sci):
        self.ID = ID
        self.name = name
        self.english = eng
        self.maths = mat
        self.science = sci
        
    def studDetails(self):
        print("\nStudent Details :-")
        print("ID -",self.ID)
        print("Name -",self.name)
        print("English Grade -",self.english)
        print("Maths Grade :",self.maths)
        print("Science Grade :",self.science)
        
    def avgGrade(self):
        print("\nAverage Grade :", (self.english + self.maths + self.science)/3)
        
        
ID = int(input("Enter student ID : "))
nm = input("Enter student name : ")
print("Enter grades for : ")
m=int(input("Maths - "))
e=int(input("English - "))
s=int(input("Science - "))
stud = Student(ID, nm, e, m, s)
stud.studDetails()
stud.avgGrade()
```

#### Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class facetagram:
    
    def __init__(self, username):
        self.username = username
        self.posts = {'Killua':['Joined HunterxHunter','became a pro hunter'],'Megumi':['Joined Jujutsu Kaisen','am a jujutsu sorcerer now'], 'Shikamaru':['Joined Naruto','passed chunin exams and became a shinobi']}
        
    def addPost(self, new_post):
        self.posts[self.username].append(new_post)
        print('\nNew post added to Facetagram...')
        
    def displayPosts(self):
        print("\nYour Posts : ")
        for i in self.posts[self.username]:
            print(i)
            

name = input("Enter your username : ")
user = facetagram(name)
print("\nWelcome to your Facetagram account")
user.displayPosts()
ch=input("\nAdd a new post today (y/n)? : ")
if(ch=='y'):
    post = input("Enter your new post : ")
    user.addPost(post)
    user.displayPosts()
print("\nThanks for using Facetagram")
```

#### Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDo :
    
    def __init__(self):
        self.tasks = {'T1':['watch lecture15','completed'],'T2':['submit assignment6','15-01-2024'],'T3':['take quiz4','20-01-2024']}
        
    def addTask(self, taskID, taskName, dueDate):
        self.tasks[taskID] = [taskName,dueDate]
        print()
        print(taskID,'-',taskName,'- has been added to your ToDo list.')
    
    def markTask(self,taskID):
        self.tasks[taskID][1]='completed'
        print()
        print(taskID,'-',self.tasks[taskID][0],'- has been marked as completed in your ToDo list.')
        
    def showTasks(self):
        print("\nYour ToDo List : ")
        for i in self.tasks.keys():
            print(i,'-',self.tasks[i][0],',',self.tasks[i][1])
            
            
user=ToDo()    
user.showTasks()
ch=0
while(ch!=3):
    ch=int(input("\nOptions : \n1. Mark task as completed \n2. Add a new task \n3. Exit \nChoice : "))
    if(ch==1):
        taskID = input("\nEnter task ID : ")
        user.markTask(taskID)
        user.showTasks()
    elif(ch==2):
        print("\nEnter new task details : ")
        taskID = input("Task ID - ")
        taskName = input("Task Name - ")
        dueDate = input("Due Date - ")
        user.addTask(taskID,taskName,dueDate)
        user.showTasks()
```
